```python
file.corr()['default'].sort_values()
```

correlation analysis shows that 
debt - to - income is (0.389) is the strongest predictor of default.

employment duration (-0.283) shoes that borrowers with longer employment history are less likely to default.

credit debt (0.244) also significantly increase risk.

this suggest financial stress indicators are stronger default predictors than demographic variables like age or education.

```python
file.groupby('default')['debtinc'].describe()
```

Debt to income ratio analysis
 the avg. debt to income ratio for defaulters (14.74) is significantly higher than non defaulters (8.68)

 non defaulters median : 7.3
 defaulters median : 13.8

 this suggest that debt to income ratio is a strong predictor of default risk.

 a potential risk threshold appears around 12-13,
 above which default probability increases substantially


```python
file['risk_level'] = pd.cut(file['debtinc'], bins=[0,8,12,50], labels=['low','medium','high'])
file.groupby(['risk_level', 'default']).mean()
```

risk segmentation based on debt to income ratio 

default rate increases significantly across risk buckets :

- low risk (<8>):12% default rate
- medium risk (8-12): 23% default rate 
- high risk (>12): 47% default rate 

this shows a strong monotonic relationship between debt to income ratio and default probability .
debt - to - income ratio can be used as a primary screening variable in credit underwriting


```python
risk_table = file.groupby("risk_level")['default'].mean()
risk_table *100
```

 Probability of Default by Risk Segment

Customers were segmented based on debt-to-income ratio.

Probability of Default (PD):

- Low Risk: 12.16%
- Medium Risk: 23.53%
- High Risk: 47.23%

There is a strong monotonic increase in default probability 
as debt-to-income ratio increases.

This confirms that debt-to-income ratio is a key driver 
of credit risk and should be considered in underwriting decisions.

Customers with DTI > 12 show sharply increasing default risk.
High-risk segment (~47% PD) should require stricter underwriting.
Medium-risk customers could require additional income verification.
Low-risk customers are safer for standard loan approval.
